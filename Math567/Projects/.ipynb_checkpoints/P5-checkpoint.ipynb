```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#mu=a*h/dt
```

```python
def Upwind(dx,h):
    
    assert dx>h and dx>0, f"The spatial step size {dx}  is invalid"
    
    t=np.arange(0,1+h,h)
    x=np.arange(0,1+dx,dx)
    U=np.zeros((len(t),len(x)))
    mu=lambda x: x+1
    IC=np.zeros(U.shape[1])
    
    for i in range(len(x)):
        IC[i]=x[i]+1
    U[0,:]=IC
    
    BC=np.zeros(U.shape[0])

    for j in range(len(t)):
        BC[j]=np.exp(-t[j])
    U[:,0]=BC
    
    for n in range(len(t)-1):
        for i in range(1,len(x)):
            U[n+1,i]=U[n,i]+mu(x[i])*(h/dx)*U[n,i-1]-mu(x[i])*(h/dx)*U[n,i]
    
    
    # Exact soln
    E=np.zeros((len(t),len(x)))
    
    f1=lambda x,t:(x+1)*np.exp(-t)
    
    for i in range(len(t)):
        for j in range(len(x)):
            E[i,j]=f1(x[j],t[i])
    return U,E,x
    
```

```python
h=0.01;dx=0.1
U,E,x=Upwind(dx,h)
```

```python
ER1=[]
for i in range(U.shape[0]):
    ER1.append(max(abs(U[i,:]-E[i,:])))
```

```python
plt.figure(figsize=(16, 6), dpi=80)
for i in range(U.shape[0]):
    plt.subplot(1, 2, 1)
    plt.plot(U[i,:])
    plt.title('Approximate')
    plt.subplot(1, 2, 2)
    plt.plot(E[i,:])
    plt.title('Exact')
    
    
```

```python
pos1=np.where(x==0.5)[0][0]
pos2=np.where(x==1)[0][0]
######################

```

```python
plt.figure(figsize=(16, 6), dpi=80)
plt.subplot(1, 2, 1)
plt.plot(U[:,pos1])
plt.plot(U[:,pos2])
plt.subplot(1, 2, 2)
plt.plot(abs(U[:,pos1]-E[:,pos1]))
plt.plot(abs(U[:,pos2]-E[:,pos2]))
plt.show()
```
