```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
'Odeint Method'
'dydx=-x/y'

def model(y,x):
    dydx=-x/y
    return dydx   
```

```python
'Runge Kutta Method'

def RK2(f,y0,x):
    H=x[1]-x[0]
    N=len(x)
    M=len(y0)
    z=np.zeros((N,M))
    z[0]=y0
    
    for k in range(0,N-1):
        k1 = f(z[k],x[k])
        k2 = f(z[k]+H*(k1/2),x[k]+H/2)
        k3 = f(z[k]+H*(k2/2),x[k]+H/2)
        k4 = f(z[k]+H*k3,x[k]+H)
        z[k+1] = z[k]+H/6*(k1+2*k2+2*k3+k4)
    return z
```

```python
'f1= exact solution'
'y0=initial value condition for y'
'x in (4,5]'



f1= lambda x:-(25-x**2)**(1/2)
x=np.linspace(4.0, 5.0, num=100, endpoint=False)
y_exact=f1(x)
y0=[-3]  

y_odeint=odeint(model,y0,x)
y_Runge=RK2(model,y0,x)

y_Runge1=y_Runge.reshape(100,)
y_odeint1=y_odeint.reshape(100,)

plt.figure(figsize=(10,6))
plt.plot(x[::5],y_odeint1[::5],'+')
plt.plot(x,y_Runge1)
plt.legend(['Odeint','Runge-Kutta'])
plt.xlabel('x')
plt.ylabel('Numerical Method')
plt.title('Odeint and Runge-kutta Method')
plt.show()
```

```python
'Absolute error'

error_odeint=abs(y_exact-y_odeint1)
error_Runge=abs(y_exact-y_Runge1)
plt.figure(figsize=(8,6))

plt.plot(x[::5],error_odeint[::5],'+')
plt.plot(x,error_Runge)
plt.legend(['Odeint','Runge'])
plt.title('Aboslute error in the Runge Kutta and Odeint Methods')
plt.ylabel('Numerical Method')
plt.xlabel('x')
plt.show()
```
