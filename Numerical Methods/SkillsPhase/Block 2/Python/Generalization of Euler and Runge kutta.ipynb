```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
sym.init_printing()
from scipy.optimize import fsolve
import scipy.integrate as integrate
```

```python
def EulerMethod(f,z0,t):
    H=t[1]-t[0]
    N=len(z0)
    M=len(t)
    z=np.zeros((M,N))
    z[0]=z0
    
    for i in range(0,M-1):
        z[i+1]=z[i]+H*f(z[i],t[i])
    return z


def F(z,t):
    x,y,p=z
    dxdt=2*x +3
    dydt=2*y+4
    dpdt=x**2+1
    return np.array([dxdt,dydt,dpdt])

a=0
b=1
M=100
z0=[1,2,3]
t=np.linspace(a,b,M)
z1=EulerMethod(F,z0,t)
x=z1[:,0]
y=z1[:,1]
p=z1[:,2]
# plt.subplot(1,3,1)
plt.plot(t,x,label='x')
# plt.subplot(1,3,1)
# plt.plot(t,y)
# plt.legend(['x','y'])
```
