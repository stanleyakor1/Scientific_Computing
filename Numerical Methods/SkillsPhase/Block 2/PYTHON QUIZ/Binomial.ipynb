```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
```

```python
k=5
N=100
P=0.5
n=3
X=np.random.binomial(n,P,N)
X
```

```python
plt.figure()
fig, axes= plt.subplots(1,5,figsize=(20,10))

def axess(axes):
    for i in range(0,5):
        ax=sns.distplot(X,kde=True,ax=axes.flatten()[i])
    return ax
axess(axes)
plt.show()
```
