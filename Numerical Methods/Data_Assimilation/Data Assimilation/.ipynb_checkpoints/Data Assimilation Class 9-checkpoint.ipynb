```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def h(t):
    if 1<=t and t<=10:
        return 1
    elif 11<=t and t<=20:
        return 0.1
    elif 21<=t:
        return 1
```

```python
T=30
```

```python
t=np.linspace(0,T,T+1)
n=len(t)
x=np.zeros(n)
x[0]=np.random.normal(0,1,1)
y=np.zeros(n)
for i in range(1,n):
    x[i]=0.8*x[i-1]+np.random.normal(0,0.4)
    y[i]=h(t[i])*x[i]+np.random.normal(0,0.1)
```

```python
plt.plot(t,x)
```

```python
plt.plot(t[1:],x[1:])
#plt.plot(t[1:],y[1:])
plt.plot(t[1:],y[1:],"+")
plt.legend(["True State","Data Model"])
plt.show()
```

# EXTENDED KALMAN

```python
T=100
t=np.linspace(1,T,T)
n=len(t)
x=np.zeros(n)
y=np.zeros(n)
x[0]=np.random.normal(0.1,0.01)
error=np.random.normal(0,0.05)
for i in range(1,n):
    x[i]=0.9*x[i-1]*(1-x[i-1])+ error
    y[i]=x[i]+np.random.normal(0,0.05)
```

```python
plt.plot(t,y)
```

```python
def kalman_filter(sigma,mean,r2,q2,alpha,H):
    gain_matrix=np.zeros(len(t))
    mean_new=np.zeros(len(t))
    sigma_new=np.zeros(len(t))
    mean_old=alpha*mean
    sigma_old=q2+alpha*sigma*alpha
    for i in range(1,len(t)):
        gain_matrix[i]=sigma_old*H*(H*sigma_old*H+r2)**(-1)
        sigma_new[i]=(1-gain_matrix[i]*H)*sigma_old
        mean_new[i]=mean_old+gain_matrix[i]*(y[i]-H*mean_old)
        mean_old=alpha*mean_new[i]
        sigma_old=q2+(alpha**2)*sigma_new[i]
    return sigma_new, mean_new, gain_matrix
```

```python
sigma=0.1
mean=0
r2=0.1
q2=0.5
alpha=0.7
H=1
Sigma_new,Mu_new,gain_Matrix=kalman_filter(sigma,mean,r2,q2,alpha,H)
```
