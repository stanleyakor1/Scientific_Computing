```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt=10**(21)
cp=1.05*10**(23)
alpha=0.3
e=0.61
s=1372
sigma=5.67*10**(-8)
t=np.linspace(0,1000*dt,1001)
T=np.zeros(len(t))
T[0]=0
for i in range(0,len(t)-1):
    T[i+1]=T[i]+ dt/cp*(s*(1-alpha))-dt/cp*(4*e*sigma*(T[i])**4)
```

```python
plt.plot(t[1:10],T[1:10])
```

```python
mean_T=np.mean(T)
```

```python
T_anom=T-mean_T
```

```python
plt.plot(T_anom[1:10])
```

# Process Model

```python
x=np.zeros(3)
x[0]=0.2
for i in range(0,len(x)-1):
    x[i+1]=4*x[i]*(1-x[i])
```

```python
print(x)
```

```python
error=np.random.normal(0,0.001,3)
error2=np.random.normal(0,0.9,3)
```

```python
y=x+error
```

```python
y2=x+error2
```

```python
plt.plot(y,"b*",markersize=7)
```

```python
x2=np.linspace(-0.1,1.1,100)
J=(x2-y[1])**2+(4*x2*(1-x2)-y[2])**2
J1=(x2-y2[1])**2+(4*x2*(1-x2)-y2[2])**2
```

```python
plt.plot(x2,J)
plt.plot(x2,J1)
```

# Stochastic

```python
def h_t(t):
    if 1<=t<=10 or 21<=t<=30:
        return 1
    elif 11<=t<=20:
        return 0.1
```

```python
h_t(22)
```

```python
T=30
t=np.linspace(0,T,T+1)
n=len(t)
x=np.zeros(n)
y=np.zeros(n)
x[0]=np.random.normal(0,1)
error=np.random.normal(0,0.4)
for i in range(1,n):
    x[i]=0.8*x[i-1]+ error
    y[i]=h_t(t[i])*x[i]+np.random.normal(0,0.1)
```

```python
plt.plot(t[1:],y[1:])
```
