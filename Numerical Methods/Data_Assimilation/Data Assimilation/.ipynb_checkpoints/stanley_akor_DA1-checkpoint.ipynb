```python
import sympy as sp # Import package SymPy
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.Symbol ( 'x') # symbolic variable defined
f = 32*x ** 3 - 48 * x ** 2 + 17.08 * x - 0.54 # defined in claim Cubic Equation Solutions
x = sp.solve (f) # call the function solve solve equations
x 
```

```python
def Cubic(a,b,N):
    x=np.linspace(a,b,N)
    n=len(x)
    y=np.zeros(n)
    for i in range(0,n):
        y[i]=32*x[i] ** 3 - 48 * x[i] ** 2 + 17.08 * x[i] - 0.54
    return y,x
```

```python
a=-10
b=10
N=100
y,x=Cubic(a,b,N)
```

```python
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
