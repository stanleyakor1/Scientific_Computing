```python
def Zero(alpha,e):
    t=np.zeros(1001)
    T=np.zeros(1001)
    for i in range(0, len(t)-1):
        t[i+1]=i*dt
        T[i+1]=T[i]+ (dt/cp)*(S*(1-alpha))-(dt/cp)*4*e*delta*(T[i])**(4)
    return t,T
```

```python
alpha=0.3
e=0.61
t,T=Zero(alpha,e)
```

```python
t
```

```python
plt.plot(t,T)
plt.title("Zero Energy Balance Model")
plt.ylabel("Temp(K)")
plt.xlabel("t(sec)")
plt.show()
```

```python
plt.plot(t[1:10],T[1:10])
plt.title("Zero Energy Balance Model")
plt.ylabel("Temp(K)")
plt.xlabel("t(sec)")
plt.show()
```

```python
T_mean=np.mean(T)
T_anom=T-T_mean
```

```python
plt.plot(t,T_anom)
plt.title("Zero Energy Balance Model")
plt.ylabel("Temp(K)")
plt.xlabel("t(sec)")
plt.show()
```

```python
plt.plot(t[1:10],T_anom[1:10])
plt.title("Zero Energy Balance Model")
plt.ylabel("Temp(K)")
plt.xlabel("t(sec)")
plt.show()
```

```python
def Zeroth(alpha,e,Cp,gamma,s):
    t=np.zeros(1001)
    T=np.zeros(1001)
    dt=1*10**(21)
    for i in range(0, len(t)-1):
        t[i+1]=i*dt
        T[i+1]=T[i]+ (dt/Cp)*(s*(1-alpha))-(dt/Cp)*4*e*gamma*(T[i])**(4)
    return t,T
```

```python
s=111
gamma=5.67*10**(-10)
Cp=1.05*10**(5)
alpha=0.3
e=0.61
t,T=Zeroth(alpha,e,Cp,gamma,s)
```

```python
T_mean=np.mean(T)
T_anom=T-T_mean
```

```python
plt.plot(t,T_anom)
```
