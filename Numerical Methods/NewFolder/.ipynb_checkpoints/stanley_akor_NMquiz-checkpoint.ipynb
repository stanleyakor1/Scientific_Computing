```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt=0.1;T=1;k=0.5;a=-1;b=1;mu=12.5;dx=np.sqrt((dt*k)/mu);Nx=int((b-a)/dx);Nt=int(T/dt);L=(b-a/2);UTa=0;UTb=0
```

```python
A=np.zeros((Nx-1,Nx-1))
x=np.arange(a,b,dx)

for i in range(Nx-2):
    A[i,i]=A[Nx-2,Nx-2]=1+2*mu
    A[i+1,i]=A[i,i+1]=-mu

U0=[]
for i in x:
    if -L/2<=i<=L/2:
        U0_1=1
        U0.append(U0_1)
    else:
        U0_1=0
        U0.append(U0_1)
F=np.zeros(Nx-1)
for i in range(Nx-1):
    F[i]=1+(x[i])**2
    F[0]=F[0]+U0[0]
    F[Nx-2]=F[Nx-2]+U0[Nx-2]

for i in range(Nt):
     Ui=np.matmul(np.linalg.inv(A),U0[1:Nx])
     U1=[UTa]+Ui.tolist()+[UTb]
     plt.plot(x,U1)
     U0=U1
plt.xlabel("x")
```
