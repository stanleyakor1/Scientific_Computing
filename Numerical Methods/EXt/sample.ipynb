```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = -2; b = 2; T = 1; dx = 0.05; K =1; mu = 0.4; dt = (mu* dx**2)/K; Nx = round((b-a)/dx); Nt = round(T/dt)
x = []; UTa = UTb = 0
for i in range (Nx +1):
    xj = a + i*dx
    x.append(xj)
U0 = [max(1 - i**2, 0) for i in x]

A = np.zeros((Nx -1, Nx-1))

for i in range (Nx-2):
    A[i,i] = A[Nx-2, Nx-2] = 1+ 2*mu
    A[i,i+1] = A[i+1,i] = -mu

plt.figure(figsize = (12, 8))
for i in range (Nt):
    Ui = np.matmul(np.linalg.inv(A),U0[1:Nx])
    Ui = [UTa] + list(Ui) + [UTb]
    plt.plot(x,Ui)
    U0 = Ui

plt.xlabel('$x$', fontsize = 'xx-large')
plt.ylabel('$t$', fontsize = 'xx-large')
```
