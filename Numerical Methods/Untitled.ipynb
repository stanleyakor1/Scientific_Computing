```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sqrt(1+x**2)
def k(x):
    return (1+x**2)**(-1)
```

```python
a=0
b=1
h=(b-a)/100
N=100
A=np.zeros((N-1,N-1))
F=np.zeros(N-1)
x=[]
for i in range(N-1):
    x1=a+(1+i)*h
    x.append(x1)
for i in range(N-2):
    A[i,i]=h**2*k(x[i])+2
    A[1+i,i]=-1
    A[i,i+1]=-1  
for i in range(N-2):
     F[i]=h**2*f(x[i])
```

```python
k
U=np.matmul(np.linalg.inv(A),F)
```

```python
plt.plot(x,U)
plt.show()
```
