```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def exact(a,b,dt):
    t=np.arange(a,b+dt,dt)
    n=len(t)
    T1=np.zeros(n)
    for i in range(0,n):
        T1[i]=289.57+10*np.exp(-0.0388*t[i])
    return T1,t
```

```python
C=85
alpha=0.3
S=1367
epsilon=0.6
sigma=5.67*10**(-8)
a=0
b=310
dt=14
T1,t=exact(a,b,dt)
```

```python
def f(T):
    return (1-alpha)*S/(4*C) - (epsilon*sigma*T**4)/C
```

```python
def Scheme(a,b,dt):
    t=np.arange(a,b+dt,dt)
    n=len(t)
    T=np.zeros(n)
    T[0]=289.57+10
    for i in range(n-1):
        k1=f(T[i])
        T_3=T[i]+(3/4)*dt*k1
        k2=f(T_3)
        T[i+1]=T[i]+(1/3)*dt*(k1+2*k2)
    return T,t
```

```python
T,t=Scheme(a,b,dt)
```

```python
plt.figure(figsize=(10,8))
plt.plot(t,T)
plt.plot(t,T1)
plt.xlabel("time(Years)")
plt.ylabel("Temperature(K)")
plt.legend(["Approximate","Exact"])
plt.show()
```
