QUESTION 3(a)

```python
def fe(x):
    if 0<x<=(np.pi): 
        return 1
    elif x%(np.pi)<(np.pi):
        return -1   
f_vec = np.vectorize(fe)

x = np.linspace(-np.pi/3, np.pi/3, 1000)    #Setting the range of x#
y = f_vec(x)                                 #Implementing the function#

plt.plot(x, y,'*')                            #Visualization#                      
plt.ylabel("y")
plt.xlabel("x")
plt.grid()
plt.show()
```

QUESTION 3(b)

```python
x=np.arange(0,np.pi,0.001)
def sumPartial(x,N):
    a = 0
    for i in range(1,N+1):
        a+= np.sin((2*i-1)*x)/(2*i-1)
    return a*4/np.pi

N1 = 5                                 #Setting the limit of the sum"
g1 = lambda x: sumPartial(x,N1)        #Implementing the partial sum function#
g1_vec = np.vectorize(g1)
p = g1_vec(x)


N2 = 10
g2 = lambda x: sumPartial(x,N2)
g2_vec = np.vectorize(g2)
p2 = g2_vec(x)

N3 = 20
g3 = lambda x: sumPartial(x,N3)
g3_vec = np.vectorize(g3)
p3 = g3_vec(x)

plt.figure(figsize=(7,6))
plt.plot(x, p)
plt.plot(x, p2)
plt.plot(x, p3)
plt.legend(['N=5','N=10','N=20'], loc='upper left')
plt.ylabel("y")
plt.xlabel("π")
plt.grid()
plt.show()

```
