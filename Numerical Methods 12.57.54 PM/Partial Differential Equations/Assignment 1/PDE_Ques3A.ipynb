```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square
```

```python
x=np.arange(-3*np.pi,3*np.pi,0.001)
y=square(x)

plt.plot(x,y, 'b.')
plt.legend(['y'])
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
def approxi(x,N):
    a = 0
    for i in range(1,N+1):
        a+= (1-(-1)**i)*np.sin(i*x)/i
    return a*2/np.pi

N = 10
g = lambda x: approxi(x,N)
g_vec = np.vectorize(g)
y1 = g_vec(x)
plt.plot(x, y1, x, y, 'r.')
plt.legend(['fourier approx','f(x)'], loc='upper left')
plt.show()
```

```python
def fe(x):
    if 0<x%(2*np.pi)<=np.pi:
        return 1
    elif x%(2*np.pi)<(2*np.pi):
        return 0
    
    
f_vec = np.vectorize(fe)

x = np.linspace(-3*np.pi, 3*np.pi, 1000)    
y = f_vec(x)
plt.plot(x, y,'*')
plt.show()
```

```python
def f(x):
    if x%(np.pi)<0.01:
        return np.nan
    elif 0<x%(2*np.pi)<np.pi:
        return 1
    elif x%(2*np.pi)<(2*np.pi):
        return -1
    
    
f_vec = np.vectorize(f)

x = np.linspace(-3*np.pi, 3*np.pi, 1000)    
y = f_vec(x)
plt.plot(x, y)
plt.show()
```
